```python
score=0
```

```python
answer=input("How many layers are there in the TCP/IP model? ")
if answer.lower()=="4":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("The OSI model consists of how many layers? ")
if answer.lower()=="7":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("The MAC address is, How much bits?: ")
if answer.lower()=="48":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("The maximum efficiency of Slotted ALOHA is: ")
if answer=="36.8":
  print("Correct!")
  score+=1
elif answer=="36.8%":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("The maximum efficiency of Pure ALOHA is: ")
if answer=="18.4":
  print("Correct!")
  score+=1
elif answer=="18.4%":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("What is the size of a standard Ethernet MAC address?(in bits) ")
if answer=="48":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("Which protocol operates at the Network Layer? ")
if answer.lower()=="ip":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("IPv4 address is how many bits long? ")
if answer.lower()=="32":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("How many bits are in an IPv6 address? ")
if answer.lower()=="128":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```

```python
answer=input("How many bits are there in each block of an IPv6 address? ")
if answer.lower()=="16":
  print("Correct!")
  score+=1
else:
  print("Incorrect!")
```
