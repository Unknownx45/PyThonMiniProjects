```python
score=0
```

```python
answer = input("You're on a dirt road, it has come to an end and you can go left or right. Which way would you like to go? ").lower()
if answer == "left":
    score += 1
    answer = input("You come to a river, you can walk around it or swim accross? Type walk to walk around and swim to swim accross: ")
    if answer == "swim":
        score += 1
        print("You swam across and were eaten by an alligator.")
        print("Game Over")
        print("Your Score of deciding the path before eliminating: ",score-1)
    elif answer == "walk":
        score += 1
        print("You walked for many miles, ran out of water and you lost the game.")
        print("Game Over")
        print("Your Score of deciding the path before eliminating: ",score-1)
    else:
        print("Not a valid option. You lose.")
        print("Game Over")
        print("Your Score of deciding the path before selecting invalid path: ",score-1)
elif answer == "right":
    score += 1
    answer = input("You come to a bridge, it looks wobbly, do you want to cross it or head back (cross/back)? ")
    if answer == "back":
        score += 1
        print("You go back and lose.")
        print("Game Over")
        print("Your Score of deciding the path before eliminating: ",score-1)
    elif answer == "cross":
        score += 1
        answer = input("You cross the bridge and meet a stranger. Do you talk to them (yes/no)? ")
        score += 1
        if answer == "yes":
            score += 1
            print("You talk to the stranger and they give you gold. You WIN!")
            print("Game Over")
            print("Your Score of deciding the path: ",score+1)
        elif answer == "no":
            score += 1
            print("You ignore the stranger and they are offended and you lose.")
            print("Game Over")
            print("Your Score of deciding the path before eliminating: ",score-1)
        else:
            print("Not a valid option. You lose.")
            print("Game Over")
            print("Your Score of deciding the path before eliminating: ",score-1)
    else:
        print("Not a valid option. You lose.")
        print("Game Over")
        print("Your Score of deciding the path before eliminating: ",score-1)
else:
    print("Not a valid option. You lose.")
    print("Game Over")
    print("Your Score of deciding the path before eliminating: ",score-1)

print("Thankyou for giving us a galance Mr/MsðŸ˜‰",name)
```
